```python
import numpy as np


A = np.array([[1,2,3],
              [4,5,6]
              ])
B = np.array([[1,2],
              [3,4],
              [5,6]
              ])
C = A @ B

print("행렬 A의 차원", A.shape)
print("행렬 B의 차원", B.shape)
print("\n내적 결과 C:\n", C)
print("결과행렬 C의 차원:",C.shape)

```
